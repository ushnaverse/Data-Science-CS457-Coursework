```python
formatted_csv = pd.read_csv('580SurveyCleanup_formatted.csv', dtype=object)
```

## Removing NULL value columns

```python
formatted_csv = formatted_csv.dropna(axis=1, how='all') #columns which has all NULL values
formatted_csv.head()
```

## Replacing Missing values

```python

#Replacing with Mode
formatted_csv['Country of Citizenship'] = formatted_csv['Country of Citizenship'].fillna(formatted_csv['Country of Citizenship'].mode()[0])
formatted_csv['Region'] = formatted_csv['Region'].fillna(formatted_csv['Region'].mode()[0])
formatted_csv['City / Town / District'] = formatted_csv['City / Town / District'].fillna(formatted_csv['City / Town / District'].mode()[0])
formatted_csv['Laptop Type'] = formatted_csv['Laptop Type'].fillna(formatted_csv['Laptop Type'].mode()[0])
formatted_csv['Are you employed full-time while attending Mason?'] = formatted_csv['Are you employed full-time while attending Mason?'].fillna(formatted_csv['Are you employed full-time while attending Mason?'].mode()[0])
formatted_csv['What is your approximate current level of programming skill in Python?'] = formatted_csv['What is your approximate current level of programming skill in Python?'].fillna(formatted_csv['What is your approximate current level of programming skill in Python?'].mode()[0])
formatted_csv['What is your approximate current level of programming skill in R?'] = formatted_csv['What is your approximate current level of programming skill in R?'].fillna(formatted_csv['What is your approximate current level of programming skill in R?'].mode()[0])
formatted_csv['What is your approximate current level of programming skill in SQL?'] = formatted_csv['What is your approximate current level of programming skill in SQL?'].fillna(formatted_csv['What is your approximate current level of programming skill in SQL?'].mode()[0])
formatted_csv['Undergraduate Degree'] = formatted_csv['Undergraduate Degree'].fillna(formatted_csv['Undergraduate Degree'].mode()[0])

formatted_csv['Expected Graduation date from Mason MS program?'] = formatted_csv['Expected Graduation date from Mason MS program?'].fillna(formatted_csv['Expected Graduation date from Mason MS program?'].mode()[0])
formatted_csv
```

##Why choose MODE for replacement?

Mode is particularly useful when dealing with categorical or nominal data, where the data points are divided into clear/different categories or classes.Filling missing values with the mode can help preserve the distribution of the categorical variable and using the mode is a straightforward and simple approach.

```python
formatted_csv = formatted_csv.replace('6l',61) #error in writing 61 as 6l
formatted_csv = formatted_csv.replace('7O',70)
formatted_csv
```

```python
formatted_csv['Height (Inches)'] = pd.to_numeric(formatted_csv['Height (Inches)'])
formatted_csv['Height (Inches)'] = formatted_csv['Height (Inches)'].fillna(formatted_csv['Height (Inches)'].mean())
formatted_csv
#column1_values = formatted_csv['Height (Inches)']
#print(column1_values)
```

## Why using MEAN here?

- float type is continous form of data so mean here will provide a reasonable estimate.
- It can reduce bias in your dataset, particularly when the missing values are missing at random or the distribution is relatively symmetrical.
- In continuous data, using the mode (which is a single value) may not be suitable because there might not be a clear mode, or the mode may not adequately represent the central tendency of the data.

```python
formatted_csv['What is your goal for learning about data analytics?'] = formatted_csv['What is your goal for learning about data analytics?'].fillna('To upskill and get a job')
```

Replacing missing/empty values for goal with a common answer using the number of answers we have got

```python
formatted_csv['Age (years)'] = formatted_csv['Age (years)'].replace('2 3',23)
formatted_csv['Age (years)'] = formatted_csv['Age (years)'].replace('1993',30)
formatted_csv
```

```python
formatted_csv['Country of Citizenship'] = formatted_csv['Country of Citizenship'].replace('USA','United States')
formatted_csv['Country of Citizenship'] = formatted_csv['Country of Citizenship'].replace('US','United States')
formatted_csv['Undergraduate Degree'] = formatted_csv['Undergraduate Degree'].replace('Yes',' ') #replaced yes with empty string
#formatted_csv #maintaining a standard
column1_values = formatted_csv['Undergraduate Degree']
print(column1_values)
```

#Part 3

```python

cleaned_csv = pd.read_csv('580SurveyCleanup_cleaned.csv')

# Define the list of question labels and the labels for 'Height (Inches)' and 'Age (years)'
question_labels = [
    'Which Section', 'Gender', 'Age (years)', 'Height (Inches)', 'Country of Citizenship', 'Undergraduate Degree',
    'Expected Graduation date from Mason MS program?', 'Laptop Type', 'Are you employed full-time while attending Mason?',
    'What is your approximate current level of programming skill in Python?', 'What is your approximate current level of programming skill in R?',
    'What is your approximate current level of programming skill in SQL?', 'What is your goal for learning about data analytics?'
]
height_label = 'Height (Inches)'
age_label = 'Age (years)'

# Initialize dictionaries to store medians and top values
medians = {}
top = {}

# Calculate medians and top values for specific columns
medians[height_label] = cleaned_csv[height_label].astype(float).median()
medians[age_label] = cleaned_csv[age_label].astype(float).median()

top[height_label] = cleaned_csv[height_label].astype(float).max()
top[age_label] = cleaned_csv[age_label].astype(float).max()

# Initialize a DataFrame to store descriptions
descriptions = pd.DataFrame()

# Calculate and append descriptions for each column
for label in question_labels:
    desc = cleaned_csv[label].describe().to_frame().T
    descriptions = descriptions.append(desc)

# Append medians and top values to the DataFrame
descriptions = descriptions.append(pd.Series(medians, name='Median'))
descriptions = descriptions.append(pd.Series(top, name='Max'))

# Reset the index for the DataFrame
descriptions.reset_index(drop=True, inplace=True)

# Print or use the 'descriptions' DataFrame as needed
print(descriptions)
```
